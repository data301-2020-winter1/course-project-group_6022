```python
import pandas as pd
import numpy as np
import pandas_profiling
import seaborn as sns
from matplotlib import pyplot as plt
import sys, os
sys.path.insert(0, os.path.abspath('..'))
path = "\\analysis\\Jonathan"
from scripts import project_functions as sc
df = sc.load_and_process(path)
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
#sns.countplot(x= 'FSA', hue = "Source of Infection", data=df)
sns.countplot(x= 'FSA', data=df,order = df['FSA'].value_counts().index,palette = 'viridis_r' )
plt.xlabel("Area Code")
plt.title("Case Counts By District (FSA)")
"""
Interesting note **M9V and M3N, the two districts with the highest case counts, are 1 district-width apart.** However, the 2 districts between them, M9M and M9L, have only ~200 cases each.
Finding out why that is would be very eye-opening.
"""

```
